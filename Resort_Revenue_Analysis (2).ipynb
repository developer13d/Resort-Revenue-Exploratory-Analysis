```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legen_lables,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in diffrent hotel',size=20)
plt.xlabel('hotel')
plt.ylabel('Number of reservation')
plt.legend(['canceled','Not canceled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort_Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City_Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('reservation_status_per_month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python

df['month'] = df['reservation_status_date'].dt.month

# Prepare grouped data
monthly_adr = df[df['is_canceled'] == 1].groupby('month')['adr'].mean().reset_index()

# Plot
plt.figure(figsize=(15, 8))
plt.title('ADR per Month (Canceled Reservations)', fontsize=30)
sns.barplot(x='month', y='adr', data=monthly_adr, palette='viridis')

plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Daily Rate', fontsize=14)
plt.legend(['adr'], fontsize=20)
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
Top_10_country=cancelled_data['country'].value_counts()[:10]  #showing country data top 10 count
plt.figure(figsize=(7,7))
plt.title('Top 10 countries with reservation canceled')
plt.pie(Top_10_country, autopct="%.2f%%", labels=Top_10_country.index)
plt.show()
```

```python
# Ensure date format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Group by date and calculate average ADR
not_cancelled_df_adr = df[df['is_canceled'] == 0].groupby('reservation_status_date')['adr'].mean().reset_index()
cancelled_df_adr = df[df['is_canceled'] == 1].groupby('reservation_status_date')['adr'].mean().reset_index()

# Sort by date for smoother plots
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.xlabel('Reservation Date')
plt.ylabel('ADR')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# only 2016 to 2017  september dataabs

# Convert to datetime if not already
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Filter the date range
start_date = '2016-01-01'
end_date = '2017-09-30'

# Group and filter for not cancelled
not_cancelled_df_adr = df[(df['is_canceled'] == 0) &
                          (df['reservation_status_date'] >= start_date) &
                          (df['reservation_status_date'] <= end_date)]\
                          .groupby('reservation_status_date')['adr'].mean().reset_index()

# Group and filter for cancelled
cancelled_df_adr = df[(df['is_canceled'] == 1) &
                      (df['reservation_status_date'] >= start_date) &
                      (df['reservation_status_date'] <= end_date)]\
                      .groupby('reservation_status_date')['adr'].mean().reset_index()

# Plot
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate (Jan 2016 â€“ Sept 2017)', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.xlabel('Reservation Date')
plt.ylabel('ADR')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
